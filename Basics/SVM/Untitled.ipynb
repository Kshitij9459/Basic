```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Frogs_MFCCs.csv')
```

```python
df.head()
```

```python
names=df.iloc[:,-4:-1]
names.head()
```

```python
values=df.iloc[:,:-4]
values.head()
```

```python

Species=names['Species']
Species=pd.get_dummies(Species,columns=['Species'])
```

```python
Y=np.array(names['Species'])
Y
```

```python
X=np.array(values)
len(X)
```

```python
model=svm.SVC(kernel='rbf',C=1000)
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
print(X_train.shape)
print(y_train.shape)
model.fit(X_train,y_train)
```

```python
ans=model.predict(X_test)
```

```python
accuracy_score(ans,y_test)
```
